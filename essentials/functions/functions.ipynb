## 2. Returning Result from a function

### 2.1 List and Function
We can send list to a function as an argument. We need to remember that anything recognized by python can be used as argument. 

```python
def sum_list(lst):
    s = 0

    for i in lst:
        s += i

    return s

# here we send list as argument to be processed by the function
sum_list([100, 200, 300])
```

```python
# If the argument does not satify the function, it will throw an error:
sum_list(4) # we pass int to that function
```

So anything other than list will be rejected by the function since the function is conditioned to process and eat list. We cann see that inside the function, it iterate i inside lst. Meaning that it expects iterable object. Even though it expects iterable object, we can pass single element as long as it is a list as follow:

```python
sum_list([4])
```

If a function can process list and produce int, can we feed int to a function to produce a list?

**NOTE**: Any entity recognizable by Python can be a function result.

```python
def strange_lst(n):
    the_list = [] # list for storing the result

    for i in range (0, n): # iterates the element in the range n
        the_list.insert(0, i)
        the_list.sort() # optional if we want to sort from the smallest

    
    return the_list

strange_lst(10)
```

### The result of a function can be easily assigned to a variable

```python
def wishes():
    return "Happy Birthday!"

w = wishes()

print(w)
```

```python
# Example 1
def wishes():
    print("My Wishes")
    return "Happy Birthday"

wishes()    # outputs: My Wishes


# Example 2
def wishes():
    print("My Wishes")
    return "Happy Birthday"

print(wishes())

# outputs: My Wishes
#          Happy Birthday
```

### Function produce list

```python
def create_list(n):
    mylist = []
    for i in range(n):
        mylist.append(i)
    return mylist

print(create_list(5))
```
